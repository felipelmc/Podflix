```python
import numpy as np
import pandas as pd
from thisamericanlife import ThisAmericanLife
```

Script para extração das transcrições dos episódios usando a biblioteca 'this-american-life'

```python
tal = ThisAmericanLife()
```

```python
df = pd.read_csv('data/tal_links.csv')
df.head(2)
```

```python
df['transcript'] = np.nan

for episode in df['episode']:
    if episode % 50 == 0:
        print(episode)

    phrase = ''
    try:
        for infos in tal.transcripts.get(episode).body_json['transcript']:
            if infos['speaker_name'] and infos['line']:
                phrase += infos['speaker_name']
                phrase += ': ' + infos['line'] + '\n'

        df.loc[df['episode'] == episode, 'transcript'] = phrase
        
    except:
        print('Episode ' + str(episode) + ' not found')
        df.loc[df['episode'] == episode, 'transcript'] = np.nan

```

```python
df['title'] = df['title'].str.replace('\n', '', regex=False)
df['description'] = df['description'].str.replace('\n', '', regex=False)
df['date_on_air'] = df['date_on_air'].str.replace('\n', '', regex=False)
df['description'] = df['description'].str.strip()
df['title'] = df['title'].str.strip()
df['date_on_air'] = df['date_on_air'].str.strip()
df.head(2)    
```

```python
df.to_csv('data/tal_links_full.csv', index = False)
```
